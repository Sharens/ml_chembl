```python
import polars as pl
import pandas as pd
import pyarrow as pa # If this imports, you're good!
RAW_PATH = 'libs/datasets/chembl_selected_ds.parquet'
df = pl.scan_parquet(RAW_PATH).collect()
df.head(5)
```

```python
dfNm = df.loc[(df['standard_units'] == 'nM') & (df['standard_value'].notnull())].copy()
```

```python
# Konwersja kolumny na float
dfNm["standard_value"] = pd.to_numeric(dfNm["standard_value"], errors="coerce")
```

```python
# Usunięcie wierszy z brakami po konwersji
dfNm = dfNm.dropna(subset=["standard_value"])
```

```python
dfNm.columns

```

'='  The activity value is exactly equal to the reported number.

'>'  The activity is greater than the reported number (e.g., IC50 > 10 µM).

'<'  The activity is less than the reported number (e.g., IC50 < 1 µM).

'>=' or '<=' Greater than or equal / less than or equal.

Sometimes you may also see ~ (approximately).


```python
import numpy as np


dfNm["pIC50"] = np.where(
    dfNm['standard_type'] == "IC50",
    -np.log10(dfNm["standard_value"] * 1e-9),
    np.nan
)

dfNm
```

```python

import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# 1. Histogram rozkładu pIC50
plt.figure(figsize=(6, 4))
sns.histplot(dfNm['pIC50'], bins=10, kde=True)
plt.title('Rozkład wartości pIC50')
plt.xlabel('pIC50')
plt.ylabel('Liczba molekuł')
plt.show()

# 2. Boxplot pIC50
plt.figure(figsize=(6, 4))
sns.boxplot(x=dfNm['pIC50'])
plt.title('Boxplot pIC50')
plt.xlabel('pIC50')
plt.show()

plt.show()
```

assay_type w ChemBL oznacza rodzaj testu biologicznego (assay), w którym mierzono aktywność molekuły. To jest kluczowa informacja, bo różne typy testów mierzą różne aspekty działania związku.
Najczęstsze wartości assay_type:

Binding – test wiązania (czy molekuła wiąże się z celem, np. receptorem).
Functional – test funkcjonalny (czy molekuła wpływa na aktywność biologiczną celu, np. hamuje enzym).
ADME – testy farmakokinetyczne (wchłanianie, dystrybucja, metabolizm, wydalanie).
Toxicity – testy toksyczności.
